```python
import os
from Bio import SeqIO
from Bio import Align
import csv

# Create pairwise aligner object
aligner = Align.PairwiseAligner()

# Set alignment parameters
aligner.mode = 'global'

with open("P_proteinA_fasta.txt") as f1, open("P_proteinB_fasta.txt") as f2:
    fasta_sequence_1 = SeqIO.parse(f1,'fasta')
    fasta_sequence_2 = SeqIO.parse(f2,'fasta')    
    
    for fasta_1, fasta_2 in zip(fasta_sequence_1, fasta_sequence_2):
        score = aligner.score(fasta_1.seq, fasta_2.seq)
        #print(f"Alignment score of Sequence #{fasta_1.id} : ", score)
        print(score)
        
```
